```python
# aplicando for ao dataframe

csv_to_df = pd.read_csv("import_csv.csv",index_col=0)
for titulo, linha in csv_to_df.iterrows():
    print(titulo) # trará o nome dos campos
    print(linha)
    print("\n")
```

```python
# aplicando for ao dataframe para imprimir uma coluna apenas
csv_to_df = pd.read_csv("import_csv.csv",index_col=0)
for titulo, linha in csv_to_df.iterrows():
    print(titulo, ":", linha["campo_texto"])# coloca entre aspas o nome da coluna
    print("\n")
```

```python
# criando uma coluna nova que trará o tamanho do dado dentro da colna campo_texto
# vale lembrar que p estruturas pequenas, ok em seguir desta forma.
# para tabelas grandes, usar a função apply (próximo exemplo)
csv_to_df = pd.read_csv("import_csv.csv",index_col=0)
for titulo, linha in csv_to_df.iterrows():
    csv_to_df.loc[titulo, "Tamanho campo"]= len(linha["campo_texto"])
    
print(csv_to_df)
```

```python
# criando uma coluna nova que trará a coluna fonte, de texto, em maiúsculo
csv_to_df = pd.read_csv("import_csv.csv",index_col=0)
for titulo, linha in csv_to_df.iterrows():
    csv_to_df.loc[titulo, "Maiusculo"]= linha["campo_texto"].upper()
    
print(csv_to_df)
```

```python
# criando uma coluna nova que trará o tamanho do dado dentro da colna campo_texto
# para tabelas grandes, usar a função apply
csv_to_df = pd.read_csv("import_csv.csv",index_col=0)
csv_to_df["Tamanho campo"]= csv_to_df["campo_texto"].apply(len)
    
print(csv_to_df)
```

```python
# criando uma coluna nova que trará a coluna fonte, de texto, em maiúsculo
# para tabelas grandes, usar a função apply
csv_to_df = pd.read_csv("import_csv.csv",index_col=0)
csv_to_df["Maiusculo"]= csv_to_df["campo_texto"].apply(str.upper)
    
print(csv_to_df)
```

```python
# conectando com um arquivo txt
nome_arquivo = 'texto.txt'
arquivo = open(nome_arquivo, mode='r') # r da palavra read, cuidado com o w, é de write (escrita)
texto = arquivo.read()
arquivo.close()
print(texto.readline())
```
