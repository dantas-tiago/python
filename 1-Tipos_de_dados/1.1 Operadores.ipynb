Operadores utilizados no python

* Operadores Aritméticos 
    * `+` (soma)
    * `-` (subtração)
    * `*` (multiplicação)
    * / (divisão)
    * % (resto)
    * // (divisão por inteiro)
    * ** (Exponenciação)
* Operadores de Comparação
    * `>` (maior que)
    * `>=` (maior ou igual que)
    * < (menor que)
    * <= (menor ou igual que)
    * == (igual)
    * != (diferente)
* Operadores Lógicos
    * and (E, deve acontece ao mesmo tempo)
    * or (Ou, uma condição já atende)
    * not (negação do valor)

```python
# precedência: Parenteses, exponenciação, multiplicação, divisão, adição e subtração
x = 10
y = 6
z = 6.5
print('x - 4 =', x - 4) # variavel com soma inteiro
print('x * 2 =', x * 2) # variavel com multiplicação
print('x * z =', x * z)
print('x + y + z=', x + y + z) # soma de 3 variaveis
print('x + y / 2) =', x + y / 2)
print('(x + y )/ 2) =', (x + y )/ 2)
print('x % y =', x % y)
print('x / y =', x / y) # resultado será float
print('x // y =', x // y) # irá forçar o resultado ser inteiro
print('x ** z =', x ** z)
# quando tiver dúvida, use a função type
```

### Operadores de Comparação
* Lista
    * `>` (maior que)
    * `>=` (maior ou igual que)
    * < (menor que)
    * <= (menor ou igual que)
    * == (igual)
    * != (diferente)

```python
x = 10
y = 4
texto = 'string'

print('x > y:', x > y) #lê-se "x é maior do que y?"
print('x <= y:', x < (y + 2)) #lê-se "x é menor a y?"
print('x <= 7.8:', x <= 7.8) #lê-se "x é menor ou igual a 7,8"
print('texto == palavra:', texto == 'palavra') #lê-se "texto contem o valor palavra?"
print('texto == palavra:', texto != 'palavra') #lê-se "texto não contem o valor palavra?"
print('(y * 0) == False:', (y * 0) == False) # pq falso em logica booleana é 0
print('texto != Tiago:', texto != 'Tiago')
```

### Operadores Lógicos
* Lista:
    * and (E, deve acontece ao mesmo tempo)
    * or (Ou, uma condição já atende)
    * not (negação do valor)

```python
# Lógica E (and), tb chamado de tabela verdade
print("F e F:" , False and False)
print("F e V:" , False and True)
print("V e F:" , True and False)
print("V e V:" , True and True) # só será verdadeiro qdo ambas condições forem verdadeiras
```

```python
# Lógica Ou (or), tb chamado de tabela verdade
print("F ou F:" , False or False)
print("F ou V:" , False or True) # será verdadeiro qdo uma condição for verdadeira
print("V ou F:" , True or False) # será verdadeiro qdo uma condição for verdadeira
print("V ou V:" , True or True)  # será verdadeiro qdo uma condição for verdadeira
```

```python
# inverte o valor
booleano = True
not_booleano = not(booleano)
print(booleano, not_booleano)
```
