### Números inteiros e Operadores algébricos (continhas)

```python
#números inteiros somados
5+5
```

```python
#números inteiros subtração
5-2
```

```python
#números inteiros multiplicados
5*2
```

```python
#números inteiros divisão
#obs: se a divisão der um número "quebrado", será um dado tipo float
5/2
#type(5/2)
```

```python
#números inteiro - forçando resultado da divisão como inteiro
print(5//2)
print(type(5//2))
```

```python
#números inteiros potenciação
5**2
#type(5**2)
```

```python
#números inteiros resto da divisão
5%2
#type(5%2)
```

```python
# comparando números
print(1==2)
print(1==1)
print(1==1.0)
```

### Números decimais (float)

```python
print(type(5))
print(type(5.0))
```

```python
#operações seguem mesmo padrão
print(5.0+2)
print(type(5.0+2))

print(5.0-2)
print(type(5.0-2))

print(5.0*2)
print(type(5.0*2))

print(5/2.0)
print(type(5/2.0))

print(5.0**2)
print(type(5.0**2))

```

### Conversões entre números


```python
# conversão de inteiro para float
print(type(5+6))

print(type(float(5+6)))
```

```python
# conversão de float para int
print(type(5.0+6))

print(type(int(5.0+6)))
```

### Tipo Booleano (Verdadeiro ou Falso)

```python
print(True)
print(type(True))
print(type(False))
```

### Texto (String)

```python
# Número como String
print("4") #com aspas duplas
print('4') #com aspas simples
print(type('texto')) #com aspas simples
print(type("4"))  
```

```python
# quando temos um texto, o operador + funciona com um concatenador
print("4"+'4')
print(type("4"+'4'))
```

```python
# quando temos um texto, o operador - gerará um erro
print("4"-'4')
print(type("4"-'4'))
```

```python
# quando temos um texto, o operador * funciona com um repetidor
print("4"*5)
print(type("4"*5))
```

```python
# quando temos um texto, o operador / gerará um erro
print("4"/5)
print(type("4"/5))
```

```python
# uma string é inalterada
string_inalterada = "exemplo de string"
print(string_inalterada)
print(string_inalterada[0])
```

```python
# não é possível alterar uma string
print(string_inalterada[0])='A'
```

```python
#comparando strings
print("abc"=='cde')
print("abc"=="abc")
```
