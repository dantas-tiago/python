### Conversão dos tipos de Dados

```python
#conversão do tipo int para outros tipos de dados primitivos
print(15)
print(type(15))
print(type(float(15)))
print(type(bool(1)))
print(type(bool(0)))
print(type(str(500)))
```

```python
#conversão do tipo float
print(15.5)
print(type(15.5))
print(type(int(15.5)))
print(type(bool(15.5)))
print(type(str(15.5)))
```

```python
#conversão do tipo bool
print(True)
print(type(True))
print(type(int(True)))
print(type(float(True)))
print(type(str(False)))
```

```python
#conversão do tipo str
print("15.5")
print(type("15.5"))
print(int('15'))
print(float('15.5'))
print(bool('15.5'))
print(bool(''))
```

```python
# para impressão de número é necessário converter
idade = 45
print(idade)

#saida = 'Eu tenho ' + idade + ' anos'
saida = 'Eu tenho ' + str(idade) + ' anos'
print(saida)
```

### Formatação utilizando f-strings (muito usado no pyspark)



```python
var = 15
f'Usando f-string com um valor {var}'
```

```python
var = 15
f'Usando f-string com um valor {var} e uma equação {var * 2}'
```

```python
# pode-se usar com os métodos também (mais detalhes sobre métodos a seguir)
var = 'tiago'
print(f'A variavel {var.upper()} tem {len(var)} letras')
```

```python
# usando f-string com número de casas decimais
alunos = 11
aprovados = 7
indice_aprovacao = (aprovados/alunos) * 100
print(f'O indice de aprovados foi: {indice_aprovacao:.10f}%')
```

```python
# usando f-string com número de casas decimais com %
alunos = 11
aprovados = 7
indice_aprovacao = (aprovados/alunos)
print(f'O indice de aprovados foi: {indice_aprovacao:.3%}')
```
