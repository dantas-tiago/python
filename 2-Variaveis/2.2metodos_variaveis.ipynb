- Quando criamos variáveis no python, podemos utilizar alguns métodos existentes (variam conforme tipo do dado armazenado), vamos trazer algumas para famialiarização;
- Aproveitando, existem várias funções que podemos utilizar para manipular as variáveis. Para visualização de todas, acessar o link: https://docs.python.org/3/library/functions.html


```python
## variável numérica int
var_num_int = 1985
var_num_int_negativo = -1985
```

```python
# para acessar, utilizar nome_variavel + .
print(var_num_int.bit_length()) #indica o número de bits necessários para representar esta variável, sem considerar o sinal
print(var_num_int_negativo.bit_length())
print(bin(var_num_int)) # mostra o número em binário
print(bin(var_num_int_negativo)) #bin é uma função nativa, assim como a função print()
print(hex(var_num_int)) # mostra o número em hexadecimal
```

```python
# Função de Potência.  Pow é uma função nativa
pow(5,2) #lê-se 5 elevado ao quadrado
```

```python
# Retorna o valor absoluto (valor positivo). Abs é uma função nativa
abs(var_num_int_negativo)
```

```python
# Retorna o valor com arredondamento para número decimal
round(6.7895422 ,4) # separador de decimal sendo .
```

### Utilizando variável com string


```python
# métodos e funções para variável tipo texto
var_nome = "variável string!"
print(var_nome)
```

```python
# declaração simultanea de forma múltipla
var_1, var_2 = 'nome 1', 'nome 2'
print(var_1)
print(var_2)
```

```python
# deixando a primeira letra maiúscula
var_nome.capitalize()
```

```python
# conta quantas vezes um determinado caractere repetiu dentro da variável string
var_nome.count("v")
```

```python
# transforma tudo em maiúsculo
print(var_nome.upper()) # não altera a variável
print(var_nome) # comprovação
```

```python
# transforma tudo em minúsculo
var_nome_maiusculo = var_nome.upper() # salvando o texto em maiúsculo
print(var_nome_maiusculo)
var_nome_maiusculo.lower()
```

```python
# Separando o texto, deve passar como parâmetro o caractere desejado
print(var_nome.split("i")) # cria uma lista com o resultado e consome o caracter utilizado como referência. 
#Obs.: Lista começa e termina com [ ]
print(var_nome.split()) # parâmetro padrão para split é espaço
```

```python
# encontrar uma posição de uma letra
print(var_nome.find("l")) #busca a posição da leta l
```

```python
# a função len indicará a quantidade de caracteres dentro da variável
len(var_nome)
```

```python
# usando o \n é possível fazer a quebra de linha de um texto
var_nome_com_quebra_de_linha = "usando o \npermite a quebra da linha"
print(var_nome_com_quebra_de_linha)
```

### Análises lógicas em string

```python
# fazer uma avaliação lógica se a palavra começa ou termina com um determinado caractere
print(var_nome.startswith('v')) # a palavra começa com v?
print(var_nome.startswith('V')) # a palavra começa com V? Case sensitive
print(var_nome.endswith('F')) # a palavra termina com F?
print(var_nome.endswith('!')) # a palavra termina com !?

```

```python
# verifica se o conteúdo é todo alpha numerico (letras e números)
var_teste_isalnum = "variávelstring"
print(var_nome.isalnum()) # espaço e ! não são
print(var_teste_isalnum.isalnum())
```

```python
# verifica se é minúsculo
var_teste_islower = "Variávelstring"
print(var_nome.islower()) 
print(var_teste_islower.islower())
```

### Retornando parte da palavra

```python
# posição no python começa em zero
var_nome[0]
```

```python
# posição no python começa em zero, com -1 ele pega o último caractere da direita
var_nome[-1]
```

```python
# posição no python começa em zero, para pegar um intervalo de valores, sintaxe [init:fim: passo]
print(var_nome[0:10]) # o 10 não é incluso
print(var_nome[:10])# caso não coloque nada na posição da esquerda, será considerado "do começo"
print(var_nome[2:])# caso não coloque nada na posição da direita, será considerado "até o fim"
print(var_nome[1:10]) 
print(var_nome[0:6:2])
print(var_nome[::-1]) #  - 1 no passo, inverte a palavra
```

```python
dir() # indica todas as variáveis que estão definidas. É uma função nativa.
```
