# <font color='Green'>Variáveis - Python </font>


Uma variável é como uma etiqueta (nome) que usamos para guardar informações em uma posição de memória.

- Pelo PEP8 (https://peps.python.org/pep-0008/) nomes de variáveis devem ser minúsculo e, quando for nome composto, deve ser utilizado o underscore. Ex.: nome_variavel
- Quando estiver salvando uma constante, deve ser em letras MAIÚSCULAS e, quando for nome composto, deve ser utilizado o underscore. Ex.: NOME_CONSTANTE
- Evitar nome de palavras reservadas: and, class, elif, finally, if, lambda, or, TRUE, yield, as, continue, else, for, import, None, pass, try, assert, def, except, from, in, nonlocal, raise, while, break, del, FALSE, global, is, not, return, with.

```python
# variável numérica int
var_num_int = 1
print(var_num_int)
print(type(var_num_int))
```

```python
# a Função id irá dizer em qual local da memória foi salvo esta variável.  id é uma função nativa
print(var_num_int)
print(id(var_num_int))
print(hex(id(var_num_int))) # função hex transforma o valor em hexadecimal
```

```python
# variável numérica float
var_num_float = 2.5
print(var_num_float)
print(id(var_num_float))
```

```python
var_nome = "Esta é uma variável string!"
print(var_nome)
```

```python
var_nome = "Esta é uma variável string!"
print(var_nome)
print(id(var_nome))
print(var_nome,"A posição de memória dessa variável é:", hex(id(var_nome))) 
# é possível imprimir mais de um texto no print, deve-se apenas colocar separado por ,
```
