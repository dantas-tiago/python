# <font color='Green'> Variáveis multivalor Tupla - Python </font>

Em python podemos utilizar algumas estruturas de dados que permite guardar mais de um valor dentro da variável, são eles:
- Listas (é mutável, pode variar seus valores internos)
- Set (não permite valores repetidos)
- Tuplas (imutável)
- Dicionário (mutável)

### Tuplas: alguns dos seus métodos e suas características
Características
- **Imutável**
- Utiliza parênteses ( )
- Para mais de um elemento, utilizar a vírgula como separador do elemento
- pode conter dados de tipo diferente
- são mais rápidas que as listas
- deve ser usado quando precisamos armazenar dados que não podem/devem ser alterados (maior segurança)

```python
# iniciando uma estrutura de dados de tupla
tupla = (10 , 20 , 30)
print(type(tupla))
print(tupla)
```

```python
# verificando tamanho da tupla
len(tupla)
```

```python
# contando elementos de uma tupla
tupla.count(20)
```

```python
# encontando a posição de um elemento de uma tupla
tupla.index(10)
```

```python
#mostrando elemento por posição
print(tupla[0])
print(tupla[0:2]) # lembrar que o parâmetro de "até" é não inclusivo
```

```python
# como é uma estrutura de dados imutável, não é possível adicionar elementos
tupla += [40]
```

```python
#apagar um item específico com delete, não é possível
tupla.remove(20)
```

```python
# limpar a tupla
tupla.clear()
```

```python
# deletando uma tupla
del tupla
```

```python
#chamando a tupla apagada
tupla
```

```python
# transformando a tupla em lista
tupla = (30 , 20 , 10)
print(tupla)
print(type(tupla))
tupla_em_lista = list(tupla)
print(tupla_em_lista)
print(type(tupla_em_lista))
```

```python
# agora como é lista, posso adicionar um valor
tupla_em_lista += [40]
print(tupla_em_lista)
```

```python
# convertendo de volta em tupla
print(tupla_em_lista)
tupla_nova = tuple(tupla_em_lista)
print(tupla_nova)
```

```python
# percorrendo todos os elementos da tupla  com for
indice1 = 0 
for vendo_elementos in tupla_nova:
    print(f'Este é o elemento {indice1}: {vendo_elementos} ') # para usar parametros dentro do texto, usa-se {}
    indice1 += 1
```
