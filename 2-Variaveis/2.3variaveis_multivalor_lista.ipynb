# <font color='Green'> Variáveis multivalor Lista - Python </font>

Em python podemos utilizar algumas estruturas de dados que permite guardar mais de um valor dentro da variável, são eles:
- Listas (é mutável, pode variar seus valores internos)
- Set (não permite valores repetidos)
- Tuplas (imutável)
- Dicionário (mutável)

### Listas e alguns dos seus métodos
Características
- Mutável
- Utiliza colchetes [ ]
- Para mais de um elemento, utilizar a vírgula como separador do elemento
- pode conter dados de tipo diferente
- Índices são números inteiros

```python
# criando uma lista
primeira_lista_s2 = []
print(type(primeira_lista_s2))
print(primeira_lista_s2)
```

```python
# declarando a lista, verificando o tipo de class e imprimindo os valores contidos
var_lista = [20, 30 , 15 , -70, 199]
print(type(var_lista))
var_lista
```

```python
# usando método sort para ordenar de forma crescente
var_lista.sort()
var_lista
```

```python
# usando método sort para ordenar de forma descrescente
var_lista.sort(reverse=True)
var_lista
```

```python
# invertendo uma lista
var_lista = [20, 30 , 15 , -70, 199]
print(var_lista)
print("Usando [::-1] ele imprime o vetor invertido\n", var_lista[::-1])
print("Mas não altera a posição do vetor:\n ",var_lista)
var_lista.reverse() #muda o vetor e salva
print("já o reverse, altera e salva as novas posições:\n", var_lista) 

```

```python
# método para contar a incidência de um valor específico
var_lista.count(30)
```

```python
# identificando o tamanho da lista criada, contará cada posição
len(var_lista)
```

```python
# função para trazer o valor máximo da lista
max(var_lista)
```

```python
# função para trazer o valor máximo da lista
min(var_lista)
```

```python
# fazendo a copia de uma lista completa
copia_var_lista = var_lista.copy()
copia_var_lista 
```

```python
# método para adicionar um elemento na lista, na última posição
var_lista.append(1536)
var_lista
```

```python
# outra forma de adiconar um elemento na lista
var_lista += [999]
var_lista
```

```python
# adicionando um elemento em uma posição (index) específico
# Exemplo, eu desejo inserir o valor 64, na posição 4
# ao fazer isso, ele desloca o atual elemento 4 para posição 5
print(var_lista)
var_lista.insert(4, 64)
print(var_lista)
```

```python
# remover um elemento que existir na lista
print(var_lista)
var_lista.remove(64) # se o valor inserido não existe, o resultado será um erro
print(var_lista)
```

```python
# editando um valor da lista
var_lista[2] = 555 #lembrando que o índice em python começa em 0
var_lista
```

```python
# apagando um elemento da lista pela posição (diferente da remove que busca o valor)
del var_lista[0] # apaga o item na posição 0
var_lista
```

```python
# adicionado um tipo de dado diferente do já armazenado
var_lista
var_lista.append("tiago")
var_lista
```

```python
# neste caso, não funcionará o sort
var_lista.sort()
```

```python
# retirando o último elemento da lista
var_lista.pop()
var_lista
```

```python
# limpando a lista
print(var_lista)
var_lista.clear()
print(var_lista)
```

```python
# lista homogenea com texto
var_lista_com_texto = ["tiago",'henrique',"dantas"]
print(var_lista_com_texto)
var_lista_com_texto.sort()
print(var_lista_com_texto)
```

### Salvando uma lista dentro de outra lista

```python
# listas de listas, é possível armazenar uma lista dentro da própria lista
lista_com_lista = [[1,2],3,4]
print(lista_com_lista)
len(lista_com_lista)
```

```python
#acessando as posições da lista de lista
print("Este é o elemento 1 da posição 1:", lista_com_lista[0][0])
print("Este é o elemento 2 da posição 1:", lista_com_lista[0][1])
print("Este é o elemento 1 da posição 2:", lista_com_lista[1])
print("Este é o elemento 1 da posição 3:", lista_com_lista[2])
```

```python
# alterando valores da lista de lista
print(lista_com_lista)
lista_com_lista[0][0]=10
print("Esta é a alteração do elemento 1 da posição 1:", lista_com_lista[0][0])
lista_com_lista[1]=20
print("Este é a alteração do  elemento 1 da posição 2:", lista_com_lista[1])
print(lista_com_lista)
```

### concatenando listas

```python
# concatenando listas
lst_1 = [1, 2, 3]
print(lst_1)
lst_2 = [4, 5, 6]
print(lst_2)
lst_concatenada = lst_1 + lst_2
print(lst_concatenada)
```

```python
# atenção com o append
lst_1 = [1, 2, 3]
print(lst_1)
lst_2 = [4, 5, 6]
print(lst_2)
lst_1.append(lst_2)
lst_1
```

```python
# altearnativa, utilizar comando extend, que irá adicionar todos os 
# elementos de lst_2 ao final de lst_1
lst_1 = [1, 2, 3]
print(lst_1)
lst_2 = [4, 5, 6]
print(lst_2)
lst_1.extend(lst_2)
lst_1
```

### Operador lógico in, verificando se existe um valor dentro de uma lista


```python
# verificando um número dentro da lista
lst_1 = [1, 2, 3]
print(2 in lst_1) # tem o valor 2 dentro da lis_1?
print(4 in lst_1) # tem o valor 2 dentro da lis_1?
```

```python
# para saber a posição de valor (tb chamado de index)
lst_1 = [10, 20, 30]
lst_1.index(10) #lembrar que as posições em python começam em 0
```

```python
# percorrendo todos os elementos da lista com for
indice1 = 0 
for vendo_elementos in lst_1:
    print(f'Este é o elemento {indice1}: {vendo_elementos} ') # para usar parametros dentro do texto, usa-se {}
    indice1 += 1
```

```python
# percorrendo todos os elementos da lista com for e enumerate
for indice2, vendo_elementos in enumerate(lst_1):
    print(f'Este é o elemento {indice2}: {vendo_elementos} ')
```

### Utilizando set, tipo especial de lista que não permite valores repetidos

```python
lst_1 = [10, 20, 20, 30]
set_teste_1 = set(lst_1)
print(lst_1)
print(set_teste_1)
print(type(set_teste_1))
```

```python
# operações com set
print(set_teste_1)
set_teste_2 = set([20, 30, 40])
print(set_teste_2)
print(set_teste_1.difference(set_teste_2)) #faz a diferença entre 1 e 2
print(set_teste_2.difference(set_teste_1)) #faz a diferença entre 2 e 1
print(set_teste_1.intersection(set_teste_2)) # faz a intersecção entre 1 e 2
print(set_teste_1.union(set_teste_2)) # faz a união entre 1 e 2
```
