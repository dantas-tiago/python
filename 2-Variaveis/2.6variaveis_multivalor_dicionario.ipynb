Em python podemos utilizar algumas estruturas de dados que permite guardar mais de um valor dentro da variável, são eles:
- Listas (é mutável, pode variar seus valores internos)
- Set (não permite valores repetidos)
- Tuplas (imutável)
- Dicionário (mutável)

### Dicionários: alguns dos seus métodos e suas características
Características
- Mutável
- Utiliza chaves {"chave": valor }
- Para mais de um elemento, utilizar a vírgula como separador do elemento
- pode conter dados de tipo diferente
- Índices podem ser de qualquer tipo tipo de dados

```python
# pq usamos dicionário? Usado mais para consulta, se for usar duas listas, como ficaria?
chaves = ["chave_1", "chave_2", "chave_3"]
valor_chaves = [50,150, 250]

# para achar uma valor de chave desejada, devemos encontrar a posição da chave:
posicao_chave = chaves.index("chave_2")
print(posicao_chave)

# e em seguida usar esta posição para encontrar o valor:

print(valor_chaves[posicao_chave])

```

```python
# criando um dicionário
primeiro_dicionario_s2 = {}
print(type(primeiro_dicionario_s2))
print(primeiro_dicionario_s2)
```

```python
# declarando o dicionário, verificando o tipo de class e imprimindo os valores contidos
var_dic = {"chave_1":50, "chave_2": 150}
print(type(var_dic))
var_dic
```

```python
# utilizando f-string com dicionarios
print(f'o valor contigo na chave 1: {var_dic["chave_1"]}')
print(f"o valor contigo na chave 1: {var_dic['chave_1']}")
#print(f"o valor contigo na chave 1: {var_dic["chave_1"]}") # se atentar à utilização de aspas duplas/simples duas vezes
```

```python
# verificando o tamanho do dicionário
len(var_dic)
```

```python
# acessando os valores
var_dic["chave_1"]
```

```python
# acessando os valores pela posição, gera erro
var_dic[0]
```

```python
# alterando os valores
var_dic["chave_1"] = 100
print(var_dic)
```

```python
# adicionando um valor de dicionário
var_dic["chave_3"]=200
var_dic
```

```python
# visualizando os itens
var_dic.items()
```

```python
# listando as chaves do dicionário
var_dic.keys()
```

```python
# imprimindo um valor de chave 
print(var_dic.keys())
print(var_dic["chave_2"])
```

```python
#listando valores relacionados a cada chave
var_dic.values()
```

```python
# apagando o valor do dicionários, utilizar o nome da chave
del var_dic["chave_3"]
print(var_dic)
```

```python
# fazendo a copia de um dicionário
novo_dic = var_dic.copy()
novo_dic
```

```python
# trazendo dados de outro dicionário
var_dic_1 = {"chave_1":50, "chave_2": 150}
var_dic_2 = {"chave_3":200, "chave_4": 250}
var_dic_1.update(var_dic_2)
print(var_dic_1)
```

```python
# trazendo dados de outro dicionário, não funcionará ao utilizar o + como fizemos para listas
var_dic_1 = {"chave_1":50, "chave_2": 150}
var_dic_2 = {"chave_3":200, "chave_4": 250}
var_dic_1 += var_dic_2 
print(var_dic_1)
```

```python
# salvando o valor de uma chave em um variável
var_valor_dic = var_dic.get("chave_1")
print(type(var_valor_dic))
print(var_valor_dic)
```

```python
# removendo o elemento do dicionário usando o nome da chave
var_dic["chave_3"]=200
print(var_dic)
var_dic.pop("chave_3")
print(var_dic)
```

```python
# removendo o último elemento do dicionário
var_dic["chave_3"]=200
print(var_dic)
var_dic.popitem()
print(var_dic)
```

```python
# pode armazenar dados de tamanhos, é tipos,  distintos
var_dic_3= {"chave_1":50, "chave_2": [100, 150], "chave_3": "Nome"}
print(type(var_dic_3))
print('Posição 0 é tipo: ', type(var_dic_3["chave_1"]))
print('Posição 1 é tipo: ', type(var_dic_3["chave_2"]))
print('Posição 2 é tipo: ', type(var_dic_3["chave_3"]))
var_dic_3
```

```python
# acessando valores com dicionarios utilizando lista
#var_dic_3[1][1] dará erro
var_dic_3["chave_2"][1]
```

```python
# acessando valores com dicionarios utilizando lista
print(type(var_dic_3["chave_3"]))
var_dic_3["chave_3"].upper() # se for texto, pode-se usar os métodos relacionados
```

```python
# pode armazenar um dicionário dentro de outro dicionário
var_dic_4= {"chave_1": {"sub_chave1_1": 50, "sub_chave1_2": 100},
            "chave_2": {"sub_chave2_1": 150, "sub_chave2_2":200}}

# consultando
var_dic_4["chave_1"]["sub_chave1_2"] # ao digitar as aspas, ele já mostra as opções
```

### Utilizando operador in para verificar se o existe o valor dentro do dicionário

```python
# verificando uma chave dentro do dicionário
var_dic = {"chave_1":50, "chave_2": 150}
print(var_dic)
print("chave_4" in var_dic) # tem o valor 2 dentro da lis_1?
print("chave_1" in var_dic) # tem o valor 2 dentro da lis_1?
```

```python
# pode-se usar o in junto com um comando if
#if "chave_1" in var_dic:
if "chave_5" in var_dic:
    print("chave encontrada, agora podemos alterar valores qdo necessário")
else:
    print("Não encontramos esta chave, tente outra vez....")
```

```python
# usando get para verificar se a chave está dentro do dicionário
d = {'chave1': 1, 'chave2':2}
val_1 = d.get('chave1')
val_5 = d.get('chave5', "não existe chave 5")
print(val_1)
print(val_5)
```
