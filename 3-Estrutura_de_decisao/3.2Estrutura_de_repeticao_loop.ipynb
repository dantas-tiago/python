# <font color='Green'>Loops - Python </font>


Em Python, existem duas estruturas de repetição principais, que permitem a execução repetida de um bloco de código (For e While).

Algumas situações em que as estruturas de repetição são fundamentais:

- Processamento de Dados: Ao lidar com grandes volumes de dados, é comum usar loops para iterar sobre cada elemento e realizar operações específicas em cada um deles.

- Interação com o Usuário: Loops são usados para criar interfaces interativas em que o programa aguarda entradas do usuário até que uma condição específica seja atendida.

- Cálculos Iterativos: Algoritmos que envolvem cálculos iterativos, como métodos numéricos de resolução de equações, também fazem uso de estruturas de repetição.

- Manipulação de Listas e Dados Compostos: Loops são usados para percorrer listas, dicionários e outras estruturas de dados compostos, permitindo a análise e manipulação dos elementos dentro delas.

Sintaxe:
**for** i **in** lista:  
    "o que deve ser executado"

```python
# for em listas
print("for em listas")
lista = [1, 2, 3 ,4 , 5]
for i in lista: #irá percorrer todos os itens do conjunto de dados "lista", em sequência
    print(i)
```

```python
# for de inteiros com função range
print("for de inteiros")
print("usando a classe", type(range(0)))
for i in range(0,10,1): #não imprime o último elemento. Obs.: o terceiro parâmetro da range pode ser opcional
    print(i)
```

```python
#for de lista heterogênea
lista_heterogenea = ['a','chuva', float , 40]
for i in lista_heterogenea:
    print(i)
```

```python
# for com dois parâmetros
print("for par x e y")
lista_x_y = [(1,1), (2,2), (3,3)]

for x,y in lista_x_y:
    print(x,y)
```

```python
# for com dicionários
dic = {'a': 2, 'b': 5}
print(type(dic))

for chave in dic:
    print(chave)
```

```python
# for com dicionários, imprimindo chave e valor
dic = {'a': 2, 'b': 5}
print(type(dic))

for chave,valor in dic.items(): #metodo items
    print(chave,valor)
```

```python
# imprimindo cada caracter com for
my_string = "python"
for i in my_string:
    print(i)
    #print(my_string[0:x]) + str(x)
```

```python
# usando for e if de forma conjunta
# identificando números pares da lista de números
for numero in range(0, 10):
    if numero % 2 == 0: # todo número par tem seu resto de divisão por 2 igual a 0
        print (numero)
```

```python
# palavra reservada continue
for numero in range(0, 10):
    if numero == 4: 
        continue #irá ignorar o print qdo numero igual a 4
    print (numero)
```

```python
# na estrutura while, diferente da for que começa "do começo" e estende ao tamanho da variável,
# não temos ciência de quantos loops irão acontecer.
x=0
while(x<15): #tb exige os dois pontos:
    print("Número", x)
    x = x +1
    # pode-se usar a expressão x +=1
    
```

```python
# pode-se usar a palavra reservada break para sair do loop
x=0
while(x<15): #tb exige os dois pontos:
    print("Número", x)
    x += 1
    break
```

```python
# pode-se usar também a palavra reservada else para ser o resultado "alternativo"
x=0
while(x<15): #tb exige os dois pontos:
    print("Número", x)
    x += 1
else:
    print ('fim do loop!')
```
