##### Numpy
O NumPy é uma biblioteca fundamental em Python para computação científica e análise de dados. Ele fornece suporte para arrays multidimensionais e funções matemáticas para manipulação eficiente de grandes conjuntos de dados numéricos.

- Eficiência: As operações em arrays NumPy são implementadas em C, o que proporciona uma execução rápida e eficiente.

- Manipulação de Arrays: NumPy oferece uma vasta gama de funções para criar, manipular e operar com arrays multidimensionais (algebra linear) de forma conveniente.

- Broadcasting: Permite realizar operações entre arrays de diferentes formas e tamanhos, estendendo automaticamente os arrays menores para combinar com os maiores, sem a necessidade de escrever loops explícitos.

- Funções Matemáticas Avançadas: Inclui uma grande variedade de funções matemáticas e estatísticas para facilitar o processamento e análise de dados.

- Integração com outras bibliotecas: NumPy é amplamente utilizado e é compatível com muitas outras bibliotecas populares em Python, como Pandas, SciPy e Matplotlib, o que o torna uma base sólida para muitas aplicações científicas e de análise de dados.

```python
import numpy as np
```

```python
# transformando a lista em um tipo array (vetor) do numpy
np_qtdade_pessoas = np.array(qtdade_pessoas)
np_area_do_local = np.array(area_do_local)
print(type(np_qtdade_pessoas))
```

```python
np_area_do_local.dtype
```

```python
# vendo o tamanho do vetor
np_area_do_local.shape
```

```python
# soma dos valores
somalista = 0
for i in qtdade_pessoas:
    somalista = somalista + i
print(somalista)

somavetor = np_qtdade_pessoas.sum()
print(somavetor)
```

```python
densidade_np = np_qtdade_pessoas / np_area_do_local
densidade_np
```

```python
#diferença entre lista e numpy
print(qtdade_pessoas + qtdade_pessoas) #adiciona o conjunt
print(np_qtdade_pessoas+ np_qtdade_pessoas) #soma as posições
```

```python
# subgrupos
print(qtdade_pessoas[1]) # lista
print(densidade_np[1]) # vetor
print(densidade_np)
print(densidade_np > 0.3) #varre  o vetor e procura qual posição atende o teste lógico
print(densidade_np[densidade_np>0.3]) 
```

```python
# usando o filto true e false para trazer os valores
densidade_filtrada = densidade_np > 0.3 # salva o vetor que tem os campos como verdadeiro
print(densidade_np)
print(densidade_filtrada)
print(densidade_np[densidade_filtrada]) # irá trazer somente os valores em true
```

```python
# mais de uma condição de forma simultanea
# aqui usar o and ou o or direto, irá gerar erro, então devemos utilizar:
# np.logical_and
print(np.logical_and(densidade_np >0.3, densidade_np< 0.4))
print(densidade_np)
```

```python
# mais de uma condição de forma simultanea
# aqui usar o and ou o or direto, irá gerar erro, então devemos utilizar:
# np.logical_or
print(np.logical_or(densidade_np >0.3, densidade_np< 0.4))
print(densidade_np)
```

```python
# mais de uma condição de forma simultanea
# aqui usar o and ou o or direto, irá gerar erro, então devemos utilizar:
print(np.logical_not(densidade_np >0.3))
print(densidade_np)
```

```python
#multiplicando por um valor escalar
print(densidade_np*10)
print(qtdade_pessoas*10)
```

```python
# usando método mediana para sumarizar dados
arrays_juntos = np.array([[10, 15, 20, 5],
                        [50, 32, 60, 15]])
print(arrays_juntos[0][:])
print(np.sort(arrays_juntos[0][:])) # mediana coloca os dados em sequencia
print(np.median(arrays_juntos[0][:])) #escolhe o elemento central
print(arrays_juntos[0][:])
```

```python
# encontrando correlação entre os arrays
print(arrays_juntos[0,:])
print(arrays_juntos[1,:])
np.corrcoef(arrays_juntos[0,:],arrays_juntos[1,:])
```

Usando for para Arrays

```python
# for para arrays 1D
np_qtdade_pessoas = np.array([10, 15, 20, 5])
i = 0

for valores in np_qtdade_pessoas:
    print(valores)
    i=i+1 # mostrando a iteração
    print(i)
```

```python
# for para arrays 2D
arrays_juntos = np.array([[10, 15, 20, 5],
                        [50,32,60,15]])
i = 0

for valores in arrays_juntos: # irá imprimir todo os dois arrays
    print(valores)
    i=i+1
    print(i)
```

```python
# for para arrays 2D, valores individuais
arrays_juntos = np.array([[10, 15, 20, 5],
                        [50,32,60,15]])
i = 0

for valores in np.nditer(arrays_juntos): # irá imprimir todo os dois arrays
    print(valores)
    i=i+1
    print(i)
```
