#### Criando um ambiente isolados para congelar as versões da bibliotecas usadas

As vezes os projetos desenvolvidos não funcionam em versões mais recentes.

Bibliotecas:
- virtualenv (https://virtualenv.pypa.io/en/latest/)
- virtualenvwrapper (https://virtualenvwrapper.readthedocs.io/en/latest/)

```python
# instalando as duas bibliotecas
!pip install virtualenv virtualenvwrapper
```

```python
import virtualenv, virtualenvwrapper
virtualenv.__version__
```

```python
# criando seu primeiro ambiente virtual
!python -m venv teste
```

```python
# ativando ambiente
!teste\Scripts\activate
```

```python
# desativando
!teste\Scripts\deactivate
```

```python
# excluindo
!rm -r .\teste\
```
