# <font color='Green'>Condições - Python </font>


As condições são expressões que permitem ao programa tomar decisões com base na avaliação de expressões lógicas. As estruturas de controle de fluxo, como condicionais, desempenham um papel crucial em programação. Aqui estão alguns conceitos importantes relacionados às condições em Python:
- Declarações condicionais: if, elif (abreviação de "else if") e else;
- Operadores lógicos: and, or, not e etc;
- Comparadores: == (igual a), != (diferente de), < (menor que), <= (menor ou igual a), > (maior que) e >= (maior ou igual a);
- Identação (4 espaços): O código dentro de um bloco condicional (ou qualquer outro bloco, como funções) é definido pela sua indentação, e a falta de indentação adequada resultará em erros;
- Operador ternário (especial): Python oferece uma forma concisa de expressar condicionais em uma única linha, conhecida como operador ternário.

```python
# estrutura simples de condicional 
if True: # o True aqui foi utilizado para entrar dentro da condição "se"
    print("Dentro do IF") # observar a identação de4 espaços após o if True:
```

```python
# usando if e else
condicao_a_ser_avaliada = 2 # Valor da variavel
if condicao_a_ser_avaliada == 0: #lembrar de usar dois "==" pois um "=" é atribuição
    print("Valor da variável é 0 ")
else:
    print("Valor não é igual a 0. Valor é: ", condicao_a_ser_avaliada )
```

```python
# usando if, elif (else if) e else
condicao_a_ser_avaliada = 1 # Valor da variavel
if condicao_a_ser_avaliada == 0: # se esta condição for verdadeira, entra nesse ponto
    print("Valor da variável é 0 ") 
elif condicao_a_ser_avaliada == 1: # se não entrar na condição anterior, irá verificar esta segunda condição
    print("Valor da variável é 1 ")
else: # se não atender nenhuma condição, entra aqui.
    print("Valor não é igual a 0. Valor é: ", condicao_a_ser_avaliada )
```

```python
# código ternário, expressão condicional em apenas uma linha
nota = 75  # Suponha que a nota do aluno seja 75
# sintaxe: resultado = valor_verdadeiro if condição else valor_falso
status = "Aprovado" if nota >= 60 else "Reprovado" 
print(status)
```

Testando Comparadores: == (igual a), != (diferente de), < (menor que), <= (menor ou igual a), > (maior que) e >= (maior ou igual a);

```python
if 2==2:
    print("Número iguais")
```

```python
if 2!=3: #deve-se perguntas: 2 é diferente de 3?
    print("Número diferentes") # se sim, entra aqui.
else:
    print("número iguais!") # se não for, entra aqui.
```

```python
if 2<3: #deve-se perguntas: 2 é menor do que 3?
    print("Sim") # se sim, entra aqui.
else:
    print("não") # se não for, entra aqui.
```

```python
if 3<=3: #deve-se perguntas: 2 é menor ou igual a 3?
    print("Sim") # se sim, entra aqui.
else:
    print("não") # se não for, entra aqui.
```

```python
if 3>3: #deve-se perguntas: 3 é maior do que 3?
    print("Sim") # se sim, entra aqui.
else:
    print("não") # se não for, entra aqui.
```

```python
if 3>=3: #deve-se perguntas: 3 é maior ou igual a 3?
    print("Sim") # se sim, entra aqui.
else:
    print("não") # se não for, entra aqui.
```

```python
# condições aninhadas
idade = 18
nome = 'Tiago'
#nome = 'Henrique'
if idade >= 18:
    print("é maior que 18")
    if nome == 'Tiago':
        print(nome, " pode entrar!")
    else:
        print(nome, "seu nome nao é Tiago, portanto não pode entrar!")
        
```

```python
#  operador and, as duas condições devem ocorrer ao mesmo tempo
idade = 18
nome = 'Tiago'
if idade >= 18 and nome == 'Tiago':
    print("é maior que 18 E", nome, "pode entrar!")
```

```python
#  operador or, pelo menos uma das duas condições devem ocorrer
idade = 17
nome = 'Tiago'
if idade >= 18 or nome == 'Tiago': 
    print("pelo menos um das condições foi atendida, Tiago!")
```

```python
# operador not
idade = 17
print("esta avaliação lógica é ", idade >= 18)
if not(idade >= 18):
    print("com o operador not, ela passa a ser ", not(idade >= 18)) # entrando no primeiro if por se tornar true
```

```python
#teste de valores
senha = 1234

tentativa = input("digite uma senha de 4 dígitos para avaliação:") #abre uma janela para receber um valor digitado
if tentativa == senha:
    print("Senha digitada", tentativa, "está correta!")
else:
    print("Senha digitada", tentativa, "está incorreta!")
```
